## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import numpy as np
import ipykernel
```

```python
def experimento_dados(num_experimentos=1000):
    count = 0 
    for i in range(num_experimentos):
        dado1 = np.random.randint(1, 7)
        dado2 = np.random.randint(1, 7)
        suma_dados = dado1 + dado2
        #ejercicio 1
        print(suma_dados)
        #ejercicio 2
        par = suma_dados % 2 == 0
        mayor_siete = suma_dados >= 7
        if par or mayor_siete: 
            count += 1
    return count/num_experimentos
    
print(f'La probabilidad de que sea par y mayor a 7 es de: {np.round(experimento_dados() *100, 2)}%')

```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.



```python
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# TODO

def experimento_bolas(num_experimentos=1000):
    blancas_rojas = 0
    mismo_color = 0 
    for _ in range(num_experimentos):
        tomar_bolas = [ball_box[np.random.randint(0, 60)] for _ in range(5)] 
        if tomar_bolas.count("White") == 3 and tomar_bolas.count("Red") == 2:
            blancas_rojas += 1
        if tomar_bolas.count("White") == 5 or tomar_bolas.count("Red") == 5 or tomar_bolas.count("Green") == 5:            mismo_color += 1
    p_blancas_rojas = blancas_rojas / num_experimentos
    p_mismo_color = mismo_color / num_experimentos
    print(f'Probabilidad de que salgan 3 blancas y 2 rojas es de: {np.round(p_blancas_rojas * 100, 2)}%')
    print(f'Probabilidad de que salgan del mismo color es de: {np.round(p_mismo_color * 100, 2)}%')

experimento_bolas()
```
